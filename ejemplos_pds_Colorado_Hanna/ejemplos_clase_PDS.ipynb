```python
#clase 7 y 8
#transformada z de la funci√≥n mostrada anteriormente y sacar polos y ceros de esta
import sympy as sym
z = sym.symbols('z')
n = sym.symbols('n', positive=True)
a = sym.symbols('a')
u = sym.Heaviside(n)

fn = (a**n)*u #definir funci√≥n
fn2 = (a**-n)*u
fnz = fn*(z**(-n)) 
fnz2 = fn2*(z**(-n))
fnmax=fn+fn2 #ecuaci√≥n
Fz_sum = sym.summation(fnz,(n,0,sym.oo))
Fz_eq  = Fz_sum.args[0]  
Fz = Fz_eq[0].simplify() 

Fz_sum2 = sym.summation(fnz2,(n,0,sym.oo))
Fz_eq2  = Fz_sum2.args[0]  
Fz2 = Fz_eq2[0].simplify() 

Fzmax_eq=Fz_eq+Fz_eq2

ROC = Fzmax_eq[1]  # condicion ROC
Fzmax=Fz+Fz2



[P,Q] = Fzmax.as_numer_denom()
P = sym.poly(P,z) #poly permite hacer operaciones anal√≠ticas en vez de con n√∫meros
Q = sym.poly(Q,z)
P_ceros = sym.roots(P) #sacar raices
Q_polos = sym.roots(Q)

print('f[n]: ')
sym.pprint(fnmax)
print('\n F[z] desde sumatoria:')
sym.pprint(Fzmax_eq)
print('\n F[z] simplificada')
sym.pprint(Fzmax)
print('\n ROC: ')
sym.pprint(ROC)
print('\n {Q_polos:veces}:',Q_polos)
print(' {P_ceros:veces}:',P_ceros)
```
